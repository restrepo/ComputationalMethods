```python
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt
%pylab inline
```

```python
#Defino mis parámetros y cndiciones iniciales
g=-9.8 #[m/s**2]
y0=200 #[m]
a=0 #Límite inferior del intervalo de integración
b=3 #Límite superior del intervalo de integración
```

```python
#Definimos la función a la que equivale mi EDO
def f(y, t):
    return g * t
```

```python
#Creamos t's que prtenezcan al intervalo [0,3]
t=np.arange(a, b, 0.01)
```

```python
#Integrando

y= integrate.odeint(f, y0, t)
```

```python
plt.plot(t,y)
plt.ylabel("Posición [m]")
plt.xlabel("Tiempo [s]")
plt.title("Posición vs. Tiempo: Aplicación a las EDO")
plt.show()
```
