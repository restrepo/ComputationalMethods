```python
import numpy as np
from scipy.misc import derivative
import matplotlib.pyplot as plt
%pylab inline
```

```python
def derivate(func,x0,dx=1,n=1,args=(),order=3):
    try:
        nn=np.array(x0).shape[0]
    except IndexError:
        nn=-1
    if nn>-1:
        y=[]
        for xx in x0:
             y.append(derivative(func,xx,dx=dx,n=n, order=order))
                
        y=np.array(y)   
    else:
        y=derivative(func,xx, dx=dx, n=n, order=order)
        
    return y

```

```python
def f(t):
    return -4.9*(t**2) + 14.7*t

def K(Diff):
    return 0.5*(Diff)**2

def U(t):
    return 9.8*(-4.9*(t**2) + 14.7*t)
    
```

```python
t=np.linspace(0,3)
Diff=derivate(f,t,dx=1E-6)
Diff2=derivate(f,t,dx=1E-6,n=2,order=3)
```

```python
plt.figure(figsize=[10,10])
plt.subplot(2,2,1)
plt.plot(t,f(t),'r')
plt.xlabel('Tiempo [t]')
plt.ylabel('Posicin [m]')
plt.title('Posicin vs. Tiempo')

plt.subplot(2,2,2)
plt.plot(t,Diff,'r')
plt.xlabel('Tiempo[t]')
plt.ylabel('Velocidad[m/s]')
plt.title('Velocidad vs. Tiempo')

plt.subplot(2,2,3)
plt.plot(t,Diff2,'r')
plt.xlabel('Tiempo [t]')
plt.ylabel('Aceleracin[m/s**2]')
plt.ylim(-12,-8)
plt.title('Aceleracin vs. Tiempo')


plt.subplot(2,2,4)
plt.plot(t,K(Diff),'r',label='Energa Cintica')
plt.plot(t,U(t), 'b',label='Energa Potencial')
plt.xlabel('Tiempo [t]')
plt.ylabel('Energa [J]')
plt.title('Energa vs. Tiempo')
plt.legend()




```
