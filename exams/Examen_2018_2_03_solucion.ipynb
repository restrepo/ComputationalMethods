## Elastic pendulum 

<div style="float: right;" markdown="1">
    <img src="https://upload.wikimedia.org/wikipedia/commons/4/47/2D_spring_Pendulum.gif"  width="300" height="400">

[From Wikipedia](https://en.wikipedia.org/wiki/Elastic_pendulum)

The spring has the rest length $l_0$ and can be stretched to length $x$. The angle of oscillation of the pendulum is $\theta$. 

The equations of motion are
$$
\ddot x =(l_0+x)\dot \theta^2 -\frac{k}{m}x + g\cos \theta\,,
$$
and
$$
\ddot \theta=-\frac{g}{l_0+x}\sin \theta-\frac{2\dot x}{l_0+x}\dot \theta
$$

We can define
$$
\omega=\dot\theta=\frac{d\theta}{dt}\,,\qquad    v=\dot x=\frac{dx}{dt}\,,
$$

Haciendo 
$$
U=\begin{bmatrix}
U_0\\
U_1\\
U_2\\
U_3
\end{bmatrix}=\begin{bmatrix}
v\\
\omega\\
x\\
\theta\\
\end{bmatrix}.
$$

Podemos obtener
\begin{align}
\frac{\operatorname{d}}{\operatorname{d} t} 
\begin{bmatrix} 
x\\
\theta\\
v\\
\omega\\
\end{bmatrix}=&
\begin{bmatrix}
v\\
\omega\\
(l_0+x)\omega^{2} - \frac{k}{m} x+ g\cos \theta\\
-\frac{1}{l_0+x}\begin{bmatrix} g\sin \theta + 2v\omega
\end{bmatrix}&\\
\end{bmatrix}\\
\end{align}


## 1. Coordenadas  x, y del movimiento

```python
# definicin de la funcin que da las derivadas de los parmetros por hallar
def du_dt(u0,t,L0=1,k=3.5,m=0.5,g=9.8):
    x,,v, = u0
    dudt = [v, , (L0+x)***2 - k/m*x + g*np.cos(), -(g*np.sin() + 2*v*)/(L0 + x)]
    return dudt

# definicin de constantes
n = 1000   
L0 = 1            # longitud en reposo [m]
k = 4           # constante elstica [N/m]
m = 0.5           # masa [kg]
g = 9.8           # gravedad [m/s]
u0 = [L0,0.3,0,0]  # condiciones iniciales de L,,v,

t = np.linspace(0,30,n)
#t = np.linspace(0,40,n)

# solucin de las ecuaciones diferenciales: L,,v,
u = odeint(du_dt,u0,t,args=(L0,k,m,g))

# coordenadas x, y del movimiento
x = L0*np.sin(u[:,1])+u[:,0]*np.sin(u[:,1])
y = -L0*np.cos(u[:,1])-u[:,0]*np.cos(u[:,1])

# grfico de las coordenadas
plt.plot(x,y,c='mediumslateblue',lw=0.8)
plt.title('Coordenadas del pndulo-resorte')
plt.xlabel('$x(t)$',size=13)
plt.ylabel('$y(t)$',size=13)
#plt.grid()
plt.show()
```

## 2. Espacio de configuraciones

```python
# definicin de condiciones iniciales aleatorias
n = 100
Lp = np.random.uniform(0,1,n)
p = np.random.uniform(-np.pi/2,np.pi/2,n)
vp = np.random.uniform(-1,1,n)
p = np.random.uniform(-1,1,n)

t = np.linspace(0,30,10000)

# clculo de L,,v, en las distintas condiciones iniciales
for L,,v, in zip(Lp,p,vp,p):
    u0 = [L,,v,]
    u = odeint(du_dt,u0,t,args=(L0,k,m,g))
    plt.plot(u[:,1],u[:,3],c='b',lw=0.1)
plt.title('Espacio de configuraciones del pndulo-resorte')
plt.xlabel('$(t)$',size=13)
plt.ylabel('$(t)$',size=13)
#plt.xlim(-0.5,0.5)
#plt.ylim(-0.5,0.5)
#plt.grid()
plt.show()
```
